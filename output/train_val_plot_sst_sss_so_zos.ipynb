```python
from pandas import *
data_sst = read_csv("../output/sst_history.csv")

tloss = data_sst['loss'].tolist()
vloss = data_sst['val_loss'].tolist()
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
plt.plot(tloss,linewidth=2)
plt.plot(vloss,linewidth=2)
plt.title('UNet SST loss',size=28)
plt.ylabel('loss',size=28)
plt.xlabel('epoch',size=28)
plt.xticks(size=20)
plt.yticks(size=20)
plt.legend(['train_sst','val_loss_sst'], loc='upper right')
```
